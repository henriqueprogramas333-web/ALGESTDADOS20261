```python
#9 - Leia uma temperatura e a unidade de origem usando match/case ("C" para Celsius, "F" para Fahrenheit ou "K" para Kelvin).
#Valide que temperaturas em Kelvin não sejam negativas. Calcule e exiba a temperatura convertida nas outras duas escalas.
#Use variáveis descritivas como temp_celsius, temp_fahrenheit e temp_kelvin. Use case _ para unidade inválida.
temp = float(input('Digite a temperatura: '))
unidade = input('Digite a unidade de temperatura (C, K, F): ')
x = 1
match unidade:
    case 'C':
        temp_celsius = temp
        temp_kelvin = (temp_celsius * 9/5) + 32
        temp_fahrenheit = temp_celsius + 273.15
    case 'K':
        if temp < 0:
            print('Kelvin não pode ser negativo')
            x = 0
        else:
            temp_kelvin = temp
            temp_celsius = temp_kelvin - 273.15
            temp_fahrenheit = (temp_celsius * 9/5) + 32
    case 'F':
        temp_fahrenheit = temp
        temp_celsius = (temp_fahrenheit - 32) * 5/9
        temp_kelvin = temp_celsius + 273.15
    case _:
        print("Unidade inválida. Use C, F ou K.")
        x=0
if x == 1:
    print(f'Temperatura em Celsius: {temp_celsius:.2f}°C')
    print(f'Temperatura em Kelvin: {temp_kelvin:.2f}°K')
    print(f'Temperatura em Fahrenheit: {temp_fahrenheit:.2f}°F')
```

```python
#12 - Leia o preço original de um produto e um código de cupom usando match/case ("PROMO10", "PROMO20", "PROMO30" ou "SEMCUPOM"). Com if aninhado, verifique se o preço já é menor que R$50,00
#Se sim, o desconto máximo permitido é 10%, independente do cupom.
#Exiba o cupom usado, o desconto em reais e o preço final.
preco = float(input("Digite o preço do produto: R$ "))
cupom = input("Digite o cupom: ")
desconto_percentual = 0
match cupom:
    case "PROMO10":
        desconto_percentual = 10
    case "PROMO20":
        desconto_percentual = 20
    case "PROMO30":
        desconto_percentual = 30
    case "SEMCUPOM":
        desconto_percentual = 0
    case _:
        print("Cupom inválido!")
        desconto_percentual = 0
if preco < 50:
    if desconto_percentual > 10:
        desconto_percentual = 10

desconto_reais = preco * (desconto_percentual / 100)
preco_final = preco - desconto_reais

print(f"\nCupom usado: {cupom}")
print(f"Desconto: R$ {desconto_reais:.2f}")
print(f"Preço final: R$ {preco_final:.2f}")
```

```python
#14 - Leia as quatro notas bimestrais de um aluno (0 a 10 cada). Valide todas. Calcule a média aritmética. Use elif para o conceito: média ≥ 9 → A, ≥ 7 → B, ≥ 5 → C, ≥ 3 → D, abaixo → E.
#Com if aninhado, exiba o status: A e B → "Aprovado", C → "Recuperação", D e E → "Reprovado". Exiba a média com 2 casas decimais e o conceito.
n1 = float(input("Digite a 1ª nota: "))
n2 = float(input("Digite a 2ª nota: "))
n3 = float(input("Digite a 3ª nota: "))
n4 = float(input("Digite a 4ª nota: "))
if (0 <= n1 <= 10 and
    0 <= n2 <= 10 and
    0 <= n3 <= 10 and
    0 <= n4 <= 10):
    media = (n1 + n2 + n3 + n4) / 4
    if media >= 9:
        conceito = "A"
    elif media >= 7:
        conceito = "B"
    elif media >= 5:
        conceito = "C"
    elif media >= 3:
        conceito = "D"
    else:
        conceito = "E"
    if conceito == "A" or conceito == "B":
        status = "Aprovado"
    elif conceito == "C":
        status = "Recuperação"
    else:
        status = "Reprovado"
    print(f"Média: {media:.2f}")
    print(f"Conceito: {conceito}")
    print(f"Status: {status}")
else:
    print("Erro: todas as notas devem estar entre 0 e 10.")
```

```python
#15 - Leia a hora atual (0-23) e o tipo de usuário usando match/case ("admin", "operador", "visitante"). Valide a hora.
#Para cada tipo, use if aninhado com elif para verificar o turno (manhã: 6-11, tarde: 12-17, noite: 18-23, madrugada: 0-5).
#Regras: admin acessa sempre, operador apenas manhã e tarde, visitante apenas tarde. Exiba se o acesso é permitido ou negado e o turno atual.
hora = int(input("Digite a hora atual (0 a 23): "))
usuario = input("Digite o tipo de usuário (admin, operador, visitante): ")
if 0 <= hora <= 23:
    if 6 <= hora <= 11:
        turno = "manhã"
    elif 12 <= hora <= 17:
        turno = "tarde"
    elif 18 <= hora <= 23:
        turno = "noite"
    else:
        turno = "madrugada"
    match usuario:
        case "admin":
            acesso = "permitido"

        case "operador":
            if turno == "manhã" or turno == "tarde":
                acesso = "permitido"
            else:
                acesso = "negado"
        case "visitante":
            if turno == "tarde":
                acesso = "permitido"
            else:
                acesso = "negado"
        case _:
            acesso = "negado"
            print("Tipo de usuário inválido!")
    print(f"Turno atual: {turno}")
    print(f"Acesso {acesso}")
else:
    print("Hora inválida! Digite um valor entre 0 e 23.")
```

```python
#29
# lendo os tipos e o poder do ataque
atacante = input("Digite o tipo do atacante (fogo/agua/planta/eletrico/terra): ")
defensor = input("Digite o tipo do defensor (fogo/agua/planta/eletrico/terra): ")
poder = float(input("Digite o poder base do ataque: "))

# lista de tipos validos
tipos_validos = ["fogo", "agua", "planta", "eletrico", "terra"]

# validando os tipos
if atacante not in tipos_validos:
    print("Erro: tipo do atacante invalido")
elif defensor not in tipos_validos:
    print("Erro: tipo do defensor invalido")
elif poder <= 0:
    print("Erro: poder deve ser positivo")
else:
    # determinando o multiplicador com base nas relacoes de tipo
    if atacante == "fogo":
        if defensor == "planta":
            multiplicador = 2.0
            resultado = "vantagem"
        elif defensor == "agua":
            multiplicador = 0.5
            resultado = "desvantagem"
        else:
            multiplicador = 1.0
            resultado = "neutro"

    elif atacante == "agua":
        if defensor == "fogo":
            multiplicador = 2.0
            resultado = "vantagem"
        elif defensor == "planta" or defensor == "eletrico":
            multiplicador = 0.5
            resultado = "desvantagem"
        else:
            multiplicador = 1.0
            resultado = "neutro"

    elif atacante == "planta":
        if defensor == "agua":
            multiplicador = 2.0
            resultado = "vantagem"
        elif defensor == "fogo":
            multiplicador = 0.5
            resultado = "desvantagem"
        else:
            multiplicador = 1.0
            resultado = "neutro"

    elif atacante == "eletrico":
        if defensor == "agua":
            multiplicador = 2.0
            resultado = "vantagem"
        elif defensor == "terra":
            multiplicador = 0.5
            resultado = "desvantagem"
        else:
            multiplicador = 1.0
            resultado = "neutro"

    elif atacante == "terra":
        if defensor == "eletrico":
            multiplicador = 2.0
            resultado = "vantagem"
        else:
            multiplicador = 1.0
            resultado = "neutro"

    # calculando o dano final
    dano = poder * multiplicador

    # exibindo o resultado
    print("Atacante:", atacante)
    print("Defensor:", defensor)
    print("Poder base:", poder)
    print("Resultado:", resultado, "(x" + str(multiplicador) + ")")
    print("Dano final:", dano)
```

```python
#30
# diarias base por tipo de acomodacao
diaria_standard = 200.0
diaria_luxo = 450.0
diaria_suite = 900.0

# lendo os dados
acomodacao = input("Tipo de acomodacao (standard/luxo/suite): ")
diarias = int(input("Numero de diarias: "))
fidelidade = input("Possui cartao fidelidade? (s/n): ")

# validando o numero de diarias
if diarias <= 0:
    print("Erro: numero de diarias deve ser positivo")
else:
    # identificando a diaria base
    if acomodacao == "standard":
        diaria_base = diaria_standard
    elif acomodacao == "luxo":
        diaria_base = diaria_luxo
    elif acomodacao == "suite":
        diaria_base = diaria_suite
    else:
        diaria_base = -1

    if diaria_base == -1:
        print("Erro: tipo de acomodacao invalido")
    else:
        subtotal = diaria_base * diarias

        # desconto por quantidade de diarias
        if diarias <= 3:
            desconto_qtd = 0.0
        elif diarias <= 7:
            desconto_qtd = subtotal * 0.10
        else:
            desconto_qtd = subtotal * 0.20

        total_parcial = subtotal - desconto_qtd

        # desconto adicional por fidelidade
        if fidelidade == "s":
            desconto_fidelidade = total_parcial * 0.05
        else:
            desconto_fidelidade = 0.0

        total_final = total_parcial - desconto_fidelidade

        # verificando o valor minimo
        if total_final < 100:
            total_final = 100.0
            print("Aviso: total ajustado ao minimo de R$100,00")

        # exibindo o resumo
        print(f"Acomodacao: {acomodacao}")
        print(f"Diaria base: R${diaria_base:.2f}")
        print(f"Numero de diarias: {diarias}")
        print(f"Subtotal: R${subtotal:.2f}")
        print(f"Desconto por diarias: R${desconto_qtd:.2f}")
        print(f"Desconto fidelidade: R${desconto_fidelidade:.2f}")
        print(f"Total final: R${total_final:.2f}")
```
